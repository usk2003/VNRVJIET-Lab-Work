# K-Nearest Neighbors (KNN) Algorithm on the Iris Dataset

This notebook demonstrates a K-Nearest Neighbors (KNN) classifier applied on the Iris dataset, including data exploration, preprocessing, model training, and visualization of results.

## Step 1: Importing Libraries and Dataset
We start by importing the necessary libraries and loading the Iris dataset.

```python
from sklearn.datasets import load_iris
# Load the dataset
iris = load_iris()
```

## Step 2: Exploring the Data
Let's print the first few rows of data and inspect the data type.

```python
# Display the first five samples
print("Sample Data:\n", iris.data[:5])
X = iris.data
print("\nData Type of X:", type(X))
```

### Feature Names

Print the names of the features available in the dataset.

```python
print("\nFeature Names:", iris.feature_names)
```

### Target Labels

Let's examine the target labels (the species classification for each sample).

```python
print("\nTarget Labels:", iris.target)
y = iris.target
```

### Target Names

The target names represent the species of the iris plants.

```python
print("\nTarget Names:", iris.target_names)
```

### Dataset Shape

Check the number of samples and features in the dataset.

```python
print("\nDataset Shape (Rows, Columns):", iris.data.shape)
```

## Step 3: Splitting the Data into Training and Testing Sets

We split the dataset into training and testing sets to evaluate model performance on unseen data.

```python
from sklearn.model_selection import train_test_split

# Split data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)

print("Training Set Shape:", X_train.shape)
print("Testing Set Shape:", X_test.shape)
```

## Step 4: Training the K-Nearest Neighbors Model

We now train a K-Nearest Neighbors model and experiment with different values of \( k \) to find the optimal value for our dataset.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

# Range of k values to test
k_range = range(1, 26)
scores = {}
scores_list = []

# Train the KNN model for each k value
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    # Predict and calculate accuracy
    y_pred = knn.predict(X_test)
    scores[k] = metrics.accuracy_score(y_test, y_pred)
    scores_list.append(scores[k])
```

```python
print()
print("\n",scores_list)
```

## Step 5: Visualizing the Accuracy for Different k Values

We plot the accuracy scores for each k value to identify the best one.

```python
import matplotlib.pyplot as plt

# Plotting the accuracy vs. k values
plt.figure(figsize=(10, 6))
plt.plot(k_range, scores_list, marker='o', linestyle='-')
plt.xlabel("k Value")
plt.ylabel("Testing Accuracy")
plt.title("Accuracy of KNN for Different k Values")
plt.show()
```

# Conclusion

By visualizing the accuracy for different k values, we can select the optimal \( k \) for our K-Nearest Neighbors classifier on the Iris dataset. This approach allows us to maximize the model's accuracy and effectiveness in classification tasks.

In this notebook, we explore the clustering of the Iris dataset using two popular unsupervised learning techniques: K-Means and Gaussian Mixture Models (GMM). We will compare the results based on silhouette scores and visualize the clustering outcomes.


## Step 1: Import Libraries and Load Dataset
We start by importing the necessary libraries, loading the Iris dataset, and preprocessing the data by scaling the features.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset from sklearn
iris = load_iris()
X = iris.data  # Features (sepal length, sepal width, petal length, petal width)
y = iris.target  # Ground truth labels (not used in unsupervised learning)

# Preprocess: Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Step 2: Apply Clustering Algorithms
We apply K-Means and Gaussian Mixture Model (GMM) clustering algorithms to the scaled features.

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 clusters for Iris dataset
kmeans_labels = kmeans.fit_predict(X_scaled)

# Apply Gaussian Mixture Model (EM algorithm)
gmm = GaussianMixture(n_components=3, random_state=42)  # 3 components for Iris dataset
gmm_labels = gmm.fit_predict(X_scaled)
```

## Step 3: Evaluate Clustering Quality
We calculate the silhouette scores for both clustering methods to assess the quality of the clusters.

```python
# Calculate Silhouette Scores to compare clustering quality
kmeans_silhouette = silhouette_score(X_scaled, kmeans_labels)
gmm_silhouette = silhouette_score(X_scaled, gmm_labels)

# Print Silhouette Scores
print(f"Silhouette Score for K-Means: {kmeans_silhouette:.4f}")
print(f"Silhouette Score for GMM (EM): {gmm_silhouette:.4f}")
```

## Step 4: Visualize Clustering Results
We visualize the clustering results by plotting the first two features (sepal length and sepal width) in a 2D scatter plot for both K-Means and GMM.

```python
# Visualize the clusters using a scatter plot (2D visualization)
# Using the first two features for visualization
plt.figure(figsize=(12, 6))

# Subplot 1: K-Means clustering results
plt.subplot(1, 2, 1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels, cmap='viridis', marker='o')
plt.title("K-Means Clustering")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

# Subplot 2: GMM clustering results
plt.subplot(1, 2, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=gmm_labels, cmap='viridis', marker='o')
plt.title("Gaussian Mixture Model (EM) Clustering")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.tight_layout()
plt.show()
```

# Conclusion
In this notebook, we applied two clustering algorithms, K-Means and Gaussian Mixture Models, to the Iris dataset. We evaluated the clustering quality using silhouette scores, with K-Means achieving a slightly higher score than GMM. The visualizations of both algorithms show the clustering results based on the first two features (sepal length and sepal width). These methods demonstrate the power of unsupervised learning in discovering patterns in data.
